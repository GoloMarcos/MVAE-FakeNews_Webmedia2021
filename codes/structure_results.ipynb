```python
caminho_dfs = ''
```

```python
from pathlib import Path
import numpy as np
import pandas as pd

basepath = Path(caminho)
files_in_basepath = basepath.iterdir()

dic = {}

for item in files_in_basepath:
  if item.is_file():
    
    partes = item.name.split('_')
    
    tam = len(partes)
    
    base = partes[0]
    
    percentual = partes[tam-1]
    
    if partes[1] == 'BoW':
      representacao = partes[1] + '_' + partes[2]
    else:
      representacao = partes[1]
    

    df = pd.read_csv(caminho + item.name, sep=';')

    best_f1_mean = max(df['f1-score-mean'])

    best_f1_std = float(df[df['f1-score-mean'] == best_f1_mean]['f1-score-std'].iloc[0])

    best_pre_mean = float(df[df['f1-score-mean'] == best_f1_mean]['precisao-mean'].iloc[0])

    best_pre_std = float(df[df['f1-score-mean'] == best_f1_mean]['precisao-std'].iloc[0])

    brest_rev_mean = float(df[df['f1-score-mean'] == best_f1_mean]['revocacao-mean'].iloc[0])

    brest_rev_std = float(df[df['f1-score-mean'] == best_f1_mean]['revocacao-std'].iloc[0])

    best_aucroc_mean = max(df['auc_roc-mean'])

    best_aucroc_std = float(df[df['auc_roc-mean'] == best_aucroc_mean]['auc_roc-std'].iloc[0])

    best_acuracia_mean = max(df['acuracia-mean'])

    best_acuracia_std = float(df[df['acuracia-mean'] == best_acuracia_mean]['acuracia-std'].iloc[0])


    if base not in dic:
      dic[base] = {}
      dic[base][representacao] = pd.DataFrame(columns=['percentual', 'precision-mean', 'precision-std',	'recall-mean', 'recall-std','f1-score-mean','f1-score-std',	'auc_roc-mean', 'auc_roc-std', 'acuracia-mean', 'acuracia-std'])
    elif representacao not in dic[base]:
      dic[base][representacao] = pd.DataFrame(columns=['percentual', 'precision-mean', 'precision-std',	'recall-mean', 'recall-std','f1-score-mean','f1-score-std',	'auc_roc-mean', 'auc_roc-std', 'acuracia-mean', 'acuracia-std'])

    df_bests = dic[base][representacao]
    df_bests = df_bests.append({'percentual' : percentual.replace('.csv','_%'),
                                'precision-mean' : best_pre_mean, 
                                'precision-std' : best_pre_std, 
                                'recall-mean' : brest_rev_mean, 
                                'recall-std' : brest_rev_std, 
                                'f1-score-mean' : best_f1_mean, 
                                'f1-score-std' : best_f1_std, 
                                'auc_roc-mean' : best_aucroc_mean, 
                                'auc_roc-std' : best_aucroc_std, 
                                'acuracia-mean' : best_acuracia_mean,
                                'acuracia-std' : best_acuracia_std} , 
                               ignore_index=True)
    dic[base][representacao] = df_bests
    
```

```python
for base in dic.keys():
  for representacao in dic[base].keys():
    dic[base][representacao].to_csv(caminho_dfs + base + '_' + representacao + '.csv', index = False)
```
